```python
lotus_cars = data[data['Make']=='Lotus'][['Engine HP', 'Engine Cylinders']]
```

```python
X = lotus_cars.to_numpy()
```

```python
X
```

```python
X.T
```

```python
XTX = np.matmul(X.T, X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```
